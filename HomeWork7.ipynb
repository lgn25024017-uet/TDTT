Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
arr=list(map(int,input().split()))
dic={}
for i in arr:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
y=0
for key,value in dic.items():
    if value>=key:
        y+=(value-key)
    else:
        y+=value

if y==0:
    print('mảng đẹp')
else:
    print(f'không phải mảng đẹp. phải xoá {y} số')


```

```python
def find_majority_element():
    try:
        arr=list(map(int,input().split()))
    except ValueError:
        return
    if not arr:
        return
    candidate=None
    count=0
    for num in arr:
        if count==0:
            candidate=num
            count=1
        elif num==candidate:
            count+=1
        else:
            count-=1
    actual_count=0
    for num in arr:
        if num==candidate:
            actual_count+=1
    if actual_count>len(arr)//2:
        print(candidate)
    else:
        print('không có phần tử thoả mãn')

find_majority_element()



    

```

Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không 
dùng bất cứ thuật toán sắp xếp nào


```python
def merge_sorted_arrays(arr1, arr2):
    # Bước 1: Chuẩn hóa 2 mảng về cùng dạng tăng dần (nếu chúng đang giảm dần)
    # Ta kiểm tra nếu phần tử đầu lớn hơn phần tử cuối thì đó là mảng giảm dần
    if len(arr1) > 1 and arr1[0] > arr1[-1]:
        arr1 = arr1[::-1]  # Đảo ngược mảng (không phải thuật toán sắp xếp)
        
    if len(arr2) > 1 and arr2[0] > arr2[-1]:
        arr2 = arr2[::-1]

    # Bước 2: Tiến hành gộp bằng kỹ thuật Hai con trỏ (Two Pointers)
    result = []
    i, j = 0, 0
    len1, len2 = len(arr1), len(arr2)
    
    # Duyệt qua cả hai mảng cho đến khi một trong hai mảng hết phần tử
    while i < len1 and j < len2:
        if arr1[i] <= arr2[j]:
            result.append(arr1[i])
            i += 1
        else:
            result.append(arr2[j])
            j += 1
            
    # Nếu mảng 1 vẫn còn phần tử thừa, nạp nốt vào kết quả
    while i < len1:
        result.append(arr1[i])
        i += 1
        
    # Nếu mảng 2 vẫn còn phần tử thừa, nạp nốt vào kết quả
    while j < len2:
        result.append(arr2[j])
        j += 1
        
    return result
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài >=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí 
tự cái.

```python
def Dem_ten(text):
    for dau_cau in [',','.']:
        text=text.replace(dau_cau,' ')
    text=text.split()
    dem=0
    for tu in text:
        if len(tu)<2:
            continue
        if not tu[0].isupper():
            continue
        cac_chu_con_lai=True
        for chu in tu[1:]:
            if not chu.islower() and chu.isalpha():
                cac_chu_con_lai=False
        if cac_chu_con_lai:
            dem+=1

    return dem




```

Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên

```python
def remove_adjacent_duplicates(text):
    text=text.split()
    result = [text[0]]  
    
    for i in range(1, len(text)):
        
        if text[i] != text[i - 1]:
            result.append(text[i])
            
    
    return " ".join(result)

input_str = input()
output_str= remove_adjacent_duplicates(input_str)

print("Xâu ban đầu:", input_str)
print("Xâu kết quả:", output_str)
```

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi 
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số

```python
def max_number_of_k_digits(num_str: str, k: int) -> str:
    n = len(num_str)
    # Số lượng chữ số tối đa ta được phép xóa bỏ
    to_remove = n - k
    stack = []
    
    for digit in num_str:
        # Trong khi stack không rỗng, chữ số hiện tại lớn hơn chữ số đỉnh stack,
        # và ta vẫn còn quyền được xóa chữ số (to_remove > 0)
        while stack and stack[-1] < digit and to_remove > 0:
            stack.pop()
            to_remove -= 1  # Đã tiêu tốn 1 lượt xóa
            
        stack.append(digit)
        
    # Nếu duyệt hết chuỗi mà vẫn chưa xóa đủ (do chuỗi có dạng giảm dần như 9876)
    # Ta chỉ việc cắt bỏ những phần tử thừa ở cuối stack
    if to_remove > 0:
        stack = stack[:-to_remove]
        
    # Gộp lại thành chuỗi kết quả và trả về
    return "".join(stack[:k])
```

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ
hay không

```python
def isValid(s: str) -> bool:
    stack = []
 
    mapping = {")": "(", "}": "{", "]": "["}
    
    for char in s:
        if char in mapping:
            # Lấy phần tử đỉnh stack nếu stack không rỗng, ngược lại gán ký tự rác '#'
            top_element = stack.pop() if stack else '#'
            
            # Nếu không khớp với ngoặc mở tương ứng
            if mapping[char] != top_element:
                return False
        else:
            # Nếu là ngoặc mở, đẩy vào stack
            stack.append(char)
            
    # Nếu stack rỗng tức là mọi ngoặc mở đều được đóng hợp lệ
    return len(stack) == 0
```

̀i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên 
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ 
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
def is_prime(k):
    """Hàm kiểm tra số nguyên tố cơ bản"""
    if k < 2: return False
    for i in range(2, int(k**0.5) + 1):
        if k % i == 0:
            return False
    return True

def find_longest_composite_sequence(n):
    # Tìm số nguyên tố đầu tiên lớn hơn N để làm mốc bắt đầu
    p1 = n + 1
    while not is_prime(p1):
        p1 += 1
        
    max_len = 0
    best_start = -1
    current_p = p1
    
    # Chúng ta sẽ tìm tiếp các số nguyên tố tiếp theo để đo khoảng cách
    for _ in range(100000): # Giới hạn vòng lặp tìm kiếm an toàn
        # Tìm số nguyên tố tiếp theo
        next_p = current_p + 1
        while not is_prime(next_p):
            next_p += 1
            
        # Độ dài chuỗi hợp số giữa hai số nguyên tố này
        current_len = next_p - current_p - 1
        
        # Nếu tìm thấy chuỗi dài hơn, cập nhật kết quả
        if current_len > max_len:
            max_len = current_len
            best_start = current_p + 1
            
        if max_len >= n: 
            break
            
        current_p = next_p

    result_sequence = list(range(best_start, best_start + max_len))
    return result_sequence
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ 
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
def find_longest_increasing_subarray(arr):
    # Nếu danh sách trống, trả về danh sách rỗng
    if not arr:
        return []

    max_sub = []      # Lưu dãy con dài nhất tìm thấy
    current_sub = [arr[0]]  # Lưu dãy con đang xét hiện tại

    for i in range(1, len(arr)):
        if arr[i] > arr[i - 1]:
            current_sub.append(arr[i])
        else:
            # Nếu bị ngắt quãng, kiểm tra xem dãy hiện tại có dài hơn dãy max cũ không
            if len(current_sub) > len(max_sub):
                max_sub = current_sub
            # Khởi tạo lại dãy hiện tại bắt đầu bằng phần tử ngắt quãng này
            current_sub = [arr[i]]
    if len(current_sub) > len(max_sub): 
        max_sub = current_sub
    return max_sub
arr=list(map(int,input().split()))
print(find_longest_increasing_subarray(arr))
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập 
vào từ bàn phím

```python
def find_longest_increasing_subsequence():
    try:
        arr = list(map(int, input().split()))
    except ValueError:
        print("Dữ liệu nhập vào không hợp lệ!")
        return

    if not arr:
        print("Danh sách rỗng!")
        return

    # 2. Khởi tạo các biến lưu trữ thông tin
    max_len = 1
    best_start = 0
    
    current_len = 1
    current_start = 0

    
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            # Dãy vẫn đang tăng tiếp tục
            current_len += 1
        else:
            # Dãy ngừng tăng, kiểm tra và cập nhật kết quả tối ưu
            if current_len > max_len:
                max_len = current_len
                best_start = current_start
            
            # Reset để bắt đầu dãy con mới tại vị trí i
            current_len = 1
            current_start = i

    # Kiểm tra lần cuối cho dãy con kết thúc ở tận cùng danh sách
    if current_len > max_len:
        max_len = current_len
        best_start = current_start

    # 4. Trích xuất dãy con dài nhất từ chỉ số đã lưu
    longest_subseq = arr[best_start : best_start + max_len]

    # 5. In kết quả ra màn hình
    print(f"Dãy con tăng dần liên tục dài nhất là: {longest_subseq}")
    print(f"Độ dài: {max_len}")

# Chạy chương trình
find_longest_increasing_subsequence()
```
