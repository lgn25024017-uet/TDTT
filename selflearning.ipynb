#### Trong Python, try...except là cấu trúc dùng để xử lý ngoại lệ (Exception Handling). Nói cách khác, nó giúp chương trình của bạn không bị "văng" (crash) khi gặp lỗi trong quá trình chạy.

-except ValueError:	       Chỉ bắt lỗi khi người dùng nhập sai kiểu dữ liệu (vd: nhập chữ thay vì số).
-except ZeroDivisionError:	Chỉ bắt lỗi khi chia một số cho 0.
-else:	        Chỉ chạy nếu khối try không có lỗi.
-finally:      Luôn luôn chạy, dù có lỗi hay không (thường dùng để đóng file, đóng kết nối database).

```python
try:
    # Khối lệnh bạn muốn chạy (có nguy cơ xảy ra lỗi)
    num = int(input("Nhập một số: "))
    result = 10 / num
    print(f"Kết quả: {result}")
except:
    # Khối lệnh sẽ chạy nếu có lỗi xảy ra ở khối try
    print("Đã xảy ra lỗi! Vui lòng kiểm tra lại đầu vào.")
```

```python
try:
    f = open("data.txt", "r")
    content = f.read()
    number = int(content)
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file data.txt")
except ValueError:
    print("Lỗi: Dữ liệu trong file không phải là số")
else:
    print("Đọc file thành công, giá trị là:", number)
finally:
    print("Kết thúc quá trình xử lý file.")
```
