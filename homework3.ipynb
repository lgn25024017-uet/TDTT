#### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    spaces = " " * (h - i)
    
    if i == 1:
        # Dòng đỉnh
        print(spaces + "*")
    elif i == h:
        # Dòng đáy
        print("*" * (2 * h - 1))
    else:
        # Các dòng giữa: khoảng trắng giữa = 2*i - 3
        inner_spaces = " " * (2 * i - 3)
        print(spaces + "*" + inner_spaces + "*")6
```

#### Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b

#### Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
import math
a = int(input("Nhập a: "))
can_bac_hai = math.sqrt(a)

# Kiểm tra xem căn bậc hai có phải số nguyên không
if can_bac_hai.is_integer():
    print(f"{a} là số chính phương")
else:
    print(f"{a} không phải là số chính phương")
```

#### Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python

n = int(input("Nhập vào số nguyên dương n: "))

s = 0  # Biến lưu tổng S(k)
k = 0  # Biến chạy k

while s + (k +1) < n:
    k += 1
    s += k

print(f"Giá trị k lớn nhất để S(k) < {n} là: k = {k}")
print(f"Khi đó, tổng S({k}) = {s}")
```

#### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
max_val = None
min_val = None

while True:
    n = int(input("Nhập một số nguyên (nhập -1 để dừng): "))
    
    if n == -1:
        break
    
    # Khởi tạo giá trị ban đầu cho max và min từ số hợp lệ đầu tiên
    if max_val is None:
        max_val = n
        min_val = n
    else:
        # Cập nhật max nếu n lớn hơn
        if n > max_val:
            max_val = n
        # Cập nhật min nếu n nhỏ hơn
        if n < min_val:
            min_val = n

# Kiểm tra xem người dùng có nhập số nào trước khi nhấn -1 không
if max_val is not None:
    print(f"Số lớn nhất: {max_val}")
    print(f"Số nhỏ nhất: {min_val}")
else:
    print("Bạn chưa nhập số nào hợp lệ.")
```

#### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
a = int(input("Nhập số nguyên dương A: "))

f1, f2 = 1, 1
# Nếu A nhỏ hơn 1 thì không có số nào trong dãy thỏa mãn (dãy bắt đầu từ 1)
if a < 1:
    print("Không có số thỏa mãn")
else:
    # f2 đóng vai trò là số Fibonacci hiện tại
    while f1 + f2 <= a:
        f_tiep_theo = f1 + f2
        f1 = f2
        f2 = f_tiep_theo
    
    print(f"Số Fibonacci lớn nhất không vượt quá {a} là: {f2}")
```

#### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n = int(input("Nhập số nguyên dương n: "))
count = 0
temp = n

if temp == 0:
    count = 0 # 0 không phải số nguyên tố
else:
    while temp > 0:
        chu_so = temp % 10  # Lấy chữ số cuối
        if chu_so in [2, 3, 5, 7]:
            count += 1
        temp //= 10  # Bỏ chữ số cuối

print(f"Số lượng chữ số là số nguyên tố: {count}")
```
