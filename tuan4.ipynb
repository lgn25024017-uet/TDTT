[Factorial] 
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n) 
với 0 < n < 100.


```python
def tinh_giai_thua(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

n = int(input("Nhập n (0 < n < 100): "))
if 0 < n < 100:
    print(f"{n}! = {tinh_giai_thua(n)}")
else:
    print("Vui lòng nhập n trong khoảng (0, 100)")
```

"[SumPrimesInRange] 
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import sys

def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Đọc nhiều bộ test từ input
for line in sys.stdin:
    if not line.strip(): break
    a, b = map(int, line.split())
    
    tong = 0
    for i in range(a, b + 1):
        if is_prime(i):
            tong += i
    print(tong)
```

"[LargestPrimeFactor] 
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

```python
n = int(input("Nhập n: "))

def largest_prime_factor(n):
    i = 2
    max_p = 0
    while i * i <= n:
        while n % i == 0:
            max_p = i
            n //= i
        i += 1
    
    # Nếu sau cùng n > 1 thì n chính là ước nguyên tố cuối cùng (và lớn nhất)
    if n > 1:
        max_p = n
    return max_p

print(largest_prime_factor(n))
```

ReverseAddPalindrome] 
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
 đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]

def reverse_num(n):
    return int(str(n)[::-1])

n = int(input())
steps = 0

# Nếu n đã là palindrome ngay từ đầu, có thể in ra luôn (tùy đề bài)
while not is_palindrome(n):
    n = n + reverse_num(n)
    steps += 1

print(f"{steps} {n}")
```

Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
W4A9 cách). Nếu không có, in ra (no number).

```python
n = int(input())
result = []

i = 1
while i * i <= n:
    cp = i * i
    s_cp = str(cp)
    # Kiểm tra các chữ số đôi một khác nhau
    if len(set(s_cp)) == len(s_cp):
        result.append(s_cp)
    i += 1

if result:
    print(" ".join(result))
else:
    print("no number")
```

[LongestCollatzUpToN] 
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
W4A10 nhỏ nhất."


```python
def get_collatz_length(x):
    length = 1
    while x != 1:
        if x % 2 == 0:
            x //= 2
        else:
            x = 3 * x + 1
        length += 1
    return length

n = int(input())
max_L = 0
best_x = 1

for x in range(1, n + 1):
    L = get_collatz_length(x)
    # Nếu tìm thấy độ dài lớn hơn, cập nhật x và L
    # Vì duyệt từ 1 đến n nên nếu gặp L bằng nhau, x nhỏ hơn đã được lưu trước đó
    if L > max_L:
        max_L = L
        best_x = x

print(f"{best_x} {max_L}")
```

Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

```python
n = int(input())
count = 0

if n % 2 != 0:
    print(0)
else:
    # Duyệt đến căn bậc hai của n để tối ưu tốc độ
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i % 2 == 0:
                count += 1
            # Kiểm tra ước còn lại (n/i) có phải ước chẵn không
            uoc_con_lai = n // i
            if uoc_con_lai != i and uoc_con_lai % 2 == 0:
                count += 1
    print(count)
```

Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
W4A13 Nếu có, in ra màn hình "true", ngược lại, in ra "false

```python
def tong_cac_uoc(num):
    tong = 0
    for i in range(1, num // 2 + 1):
        if num % i == 0:
            tong += i
    return tong

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

if tong_cac_uoc(a) == b and tong_cac_uoc(b) == a:
    print("true")
else:
    print("false")
```

[ChickenAndDog] 
"Vừa gà vừa chó, 
Bó lại cho tròn, 
Ba mươi sáu con, 
Một trăm chân chẵn" 
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số 
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và 
W4A15 số lượng chó. Ngược lại, in ra "invalid".

```python
tong_con = int(input("Nhập tổng số con: "))
tong_chan = int(input("Nhập tổng số chân: "))

found = False

# i là số lượng gà
for ga in range(tong_con + 1):
    cho = tong_con - ga
    if (2 * ga + 4 * cho) == tong_chan:
        print(f"{ga} {cho}")
        found = True
        break # Đã tìm thấy nghiệm phù hợp

if not found:
    print("invalid")
```
