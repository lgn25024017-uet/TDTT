[PerfectSquaresDistinctDigits] 
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number).

```python
n=int(input())
kqua=[]
i=1
while i*i<=n:
    cp=i*i
    s_cp=str(cp)
    if len(set(s_cp))==len(s_cp):
        kqua.append(s_cp)
    i+=1
if kqua:
    print(' '.join(kqua))       
else:
    print('no number')
```

Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt  quá A

```python
A=int(input())
f1,f2=1,1
if A<1:
    print('không có số thoả mãn')
else:
    while f2<=A:
        f_next=f1+f2
        f1=f2
        f2=f_next
print(f1)
```

Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi Dùng cách: for trên index; KHÔNG dùng hàm split()

```python
s = input("Nhập vào một chuỗi: ")

tu_dau_tien = ""
# Duyệt theo index
for i in range(len(s)):
    
    if s[i] == " ":
        if tu_dau_tien != "":
            break
    else:
        tu_dau_tien += s[i]

print("Từ đầu tiên là:", tu_dau_tien)
```

nhập một số nhỏ hơn 10000 in ra tất cả các số nguyên tố nhỏ hơn hoặc bằng n dùng sàng lọc erastosthen



```python
n = int(input("Nhập một số n (n < 10000): "))

if n < 2:
    print("Không có số nguyên tố nào.")
else:
    prime = [True] * (n + 1)
    prime[0] = prime[1] = False

    for p in range(2, int(n**0.5) + 1):
        # Nếu prime[p] vẫn là True, thì nó là số nguyên tố
        if prime[p] == True:
            # Loại bỏ tất cả các bội số của p bắt đầu từ p*p
            for i in range(p * p, n + 1, p):
                prime[i] = False
    print(f"Các số nguyên tố nhỏ hơn hoặc bằng {n} là:")
    for p in range(2, n + 1):
        if prime[p]:
            print(p, end=" ")
```
