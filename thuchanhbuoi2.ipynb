Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được
phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

```python
a1,b1,c1,a2,b2,a3= map(float,input().split())
tb= ((a1+b1+c1)+(a2+b2)*2+a3*3)/10
print(f'điểm trung bình là: {tb:.1f}')
```

“Chuỗi” trong python là các kí tự được xâu lại với nhau như những hạt trên một chiếc dây chuyền.
Chính vì thế “chuỗi” còn được gọi là “xâu”. Một chuỗi gồm các kí tự nằm trong dấu nháy đơn '' hoặc
nháy đôi "”.• Mỗi chuỗi (str) là một chuỗi các kí tự trong bảng mã Unicode kể từ Python 3 trở đi
(Unicode là một tiêu chuẩn mã hóa kí tự toàn cầu, cho phép biểu diễn hầu hết mọi kí tự trong tất cả
các ngôn ngữ trên thế giới, bao gồm cả kí hiệu, biểu tượng, emoji,….). Bạn có thể dùng ord() và
chr() để làm việc với mã Unicode.
Hãy nhập vào một kí tự chữ cái thường ‘a’…’z’, in ra mã Unicode của nó sau đó chuyển nó thành kí
tự hoa tương ứng, in kí tự hoa này ra màn hình.


```python
c = input()
print(f'mã unicode là: {ord(c)}')
print(f'chữ in hoa là: {chr(ord(c)-32)}')
```

tròn đến hai chữ số thập phân đằng sau dấu phẩy.

Viết chương trình nhận vào một số nguyên dương n và trả về "số sao" thứ n.
Một "số sao" là một số hình học trung tâm đại diện cho một hình lục giác có tâm (ngôi sao
sáu cánh), giống như hình dạng được sử dụng trong trò chơi cờ Trung Quốc.

```python
def tinh_so_sao(n):
    # Kiểm tra điều kiện: n phải là số nguyên dương
    if n <= 0:
        return "Vui lòng nhập số nguyên dương (n > 0)"
    
    # Áp dụng công thức toán học
    so_sao = 6 * n * (n - 1) + 1
    return so_sao

# --- Phần chạy thử (Test) ---
print(f"Số sao thứ 1 là: {tinh_so_sao(1)}")   
```

[DrawTriangle]
Sử dụng hàm print, bạn có thể in ra các dòng văn bản, các câu thông báo. Ngoài ra, với chút
nghệ thuật xếp đặt (ký tự), bạn có thể vẽ nhiều hình vẽ đẹp và thú vị. Giới hạn chính là trí
tưởng tượng của bạn.
Trong bài này, bạn được yêu cầu viết chương trình vẽ ra một tam giác cân bằng các dấu sao
trên 3 dòng. Hãy kết hợp việc sử dụng khoảng trắng với các dấu sao để in ra tam giác này.
Ví dụ, đoạn mã để in 2 dòng đầu của tam giác:
print("" * "")
print("" *** "")
Yêu cầu: In ra màn hình tam giác có 3 dòng như sau:
*
***
***** 

```python
def ve_tam_giac(n):
    # Vòng lặp cho i chạy từ 1 đến n
    for i in range(1, n + 1):
        
        # 1. Tính toán số lượng khoảng trắng và dấu sao
        so_khoang_trang = n - i
        so_sao = 2 * i - 1
        
        # 2. Tạo ra chuỗi ký tự tương ứng
        chuoi_khoang_trang = " " * so_khoang_trang
        chuoi_sao = "*" * so_sao
        
        # 3. Ghép khoảng trắng với dấu sao và in ra màn hình
        print(chuoi_khoang_trang + chuoi_sao)

# --- Chạy thử ---
print("Đây là tam giác 3 dòng:")
ve_tam_giac(3)

print("\nĐây là tam giác 5 dòng:")
ve_tam_giac(5)
```
