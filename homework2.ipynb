#### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a,b,c=map(float,input().split())
delta=b**2 - 4*a*c
def giai_pt_bac_2():
    if a == 0:
        return 'đây không phải phương trình bậc 2'
    elif delta < 0:
        return "Phương trình vô nghiệm thực"
    elif delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép x1 = x2 = {x}"
    else:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        return f"Phương trình có 2 nghiệm phân biệt: x1 = {x1}, x2 = {x2}"
giai_pt_bac_2()
```

#### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0, y0, r = map(float, input().split())
x, y = map(float, input().split())#toạ độ A

khoang_cach = (x - x0)**2 + (y - y0)**2
ban_kinh = r**2

# So sánh trực tiếp
if khoang_cach == ban_kinh:
    print("A nằm TRÊN đường tròn")
elif khoang_cach < ban_kinh:
    print("A nằm TRONG đường tròn")
else:
    print("A nằm NGOÀI đường tròn")
```
