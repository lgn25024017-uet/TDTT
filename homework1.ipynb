#### Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a = float(input("chiều rộng của hình chữ nhật: "))
b = float(input("chiều dài của hình chữ nhật: "))
dien_tich_hcn = a * b
pi = 3.14
r = a/2
dien_tich_hinh_tron = pi * (r ** 2)
dien_tich_con_lai = dien_tich_hcn - dien_tich_hinh_tron
print(f"diện tích còn để trồng cây là: {dien_tich_con_lai:.2f}")
```

#### Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
def ky_tu(c):
    if 'A' <= c <= 'Z':
        return chr(ord(c) + 32)
    elif 'a' <= c <= 'z':
        return chr(ord(c) - 32)   
    return c 
print(ky_tu('4'))
    
```

```python
def tim_chu_cai_lien_truoc():
    c=input('nhập chữ cái in hoa')
    #chuyển về chữ thường

    chu_thuong=c.lower()
    #xét th đặc biệt c='A'
    if chu_thuong=='a':
        ket_qua='z'
    else:
        ma_so_lien_truoc=ord(chu_thuong)-1
        ket_qua=chr(ma_so_lien_truoc)
    print(f'chữ cái thường liền trước {chu_thuong} là {ket_qua}')
tim_chu_cai_lien_truoc()
```

#### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math
def tam_giac(a, b, c):
 
    if a + b > c and a + c > b and b + c > a:
        p = (a + b + c) / 2
        dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
        # Nếu đúng, in ra diện tích luôn
        print(f'diện tích {dien_tich}')
    else:
        # Nếu sai, in ra câu thông báo
        print('không phải ba cạnh tam giác')
a, b, c = map(float, input().split())
tam_giac(a,b,c)
```

#### 9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
def chuyen_so_thanh_chu():
    try:
        so = int(input("Nhập một số nguyên từ 0 đến 9: "))
        
        # 2. Sử dụng cấu trúc match...case
        match so:
            case 0:
                print("Không")
            case 1:
                print("Một")
            case 2:
                print("Hai")
            case 3:
                print("Ba")
            case 4:
                print("Bốn")
            case 5:
                print("Năm")
            case 6:
                print("Sáu")
            case 7:
                print("Bảy")
            case 8:
                print("Tám")
            case 9:
                print("Chín")
            case _: # Dấu gạch dưới đại diện cho các trường hợp còn lại (default)
                print("Lỗi: Số nhập vào không nằm trong khoảng 0-9")
                
    except ValueError:
        print("Lỗi: Vui lòng nhập một con số nguyên!")
chuyen_so_thanh_chu()
```
