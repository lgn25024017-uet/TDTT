Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp 
xếp.)

```python

data_input = tuple(map(int, input("Nhập các số (cách nhau bởi khoảng trắng): ").split()))
k = int(input("Nhập số k: "))
n = len(data_input)
if n > 0:
    k = k % n  # Chỉ xoay những vòng thực sự cần thiết

    # Bước 3: Thực hiện xoay trái bằng slicing
    # data_input[k:] lấy từ vị trí k đến hết
    # data_input[:k] lấy từ đầu đến vị trí k (không bao gồm k)
    result = data_input[k:] + data_input[:k]
else:
    result = data_input


print("Tuple sau khi xoay trái:", result)
```

Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó


```python
pairs=input().split()

result_dict={}
for i in pairs:
    key,value=i.split(':')
    if key in result_dict:
        result_dict[key].append(value)
    else:
        result_dict[key]=[value]
print(result_dict)
```

Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0)

```python
lst=list(map(int,input().split()))

# Khởi tạo dictionary để đếm
# Ban đầu các giá trị đều bằng 0
counts = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}

for num in lst:
    if num > 0:
        counts['positives'] += 1
    elif num < 0:
        counts['negatives'] += 1
    else:
        counts['zeros'] += 1

print(counts)
```

Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện 
của từng chuỗi và lưu vào dictionary

```python

input_list = input().split()

# Khởi tạo dictionary trống
count_dict = {}

for string in input_list:
    # Nếu chuỗi đã có trong dict, tăng giá trị lên 1
    if string in count_dict:
        count_dict[string] += 1
    # Nếu chưa có, thêm vào dict với giá trị khởi tạo là 1
    else:
        count_dict[string] = 1

print("Kết quả đếm:", count_dict)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key

```python
def increasing_str(dict1,dict2):
    result=dict1.copy()
    for key,value in dict2.items():
        if key in result:
            result[key]+= value
        else:
            result[key]=value

    sorted_keys=sorted(result.keys())
    final_dict={k:result[k] for k in sorted_keys}

    return final_dict

def get_user_input(label):
   
    raw = input().split()
    return {item.split(':')[0]: int(item.split(':')[1]) for item in raw}

# Chạy chương trình
d1 = get_user_input("Dict 1")
d2 = get_user_input("Dict 2")

final = increasing_str(d1, d2)

print(final)



```

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
def go(lst,k):
    result=[]
    n=len(lst)
    for i in range(n):
        for j in range(i+1,n):
            if lst[i]+lst[j]==k:
                result.append((lst[i],lst[j]))
    result.sort(key=lambda x: x[0])

    return result

lst=list(map(int,input().split()))
k=int(input())
print(go(lst,k))



            
```

[Tách số chẵn và lẻ từ tuple] 
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt

```python
def split_even_odd_tuple(input_tuple):
   
    even_list = []
    odd_list = []
    for num in input_tuple:
        if num % 2 == 0:
            even_list.append(num)  
        else:
            odd_list.append(num)   

    even_tuple = tuple(even_list)
    odd_tuple = tuple(odd_list)

    return even_tuple,odd_tuple

my_tuple = (10, 3, 5, 8, 2, 7, 1)
evens,odds=split_even_odd_tuple(my_tuple)

print(evens)
print(odds)
```

[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong 
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
nhất trong số đó.

```python
nums=list(map(int,input().split()))
count={}
for i in nums:
    if i in count:
        count[i]+=1
    else:
        count[i]=1

#tìm key của value lớn nhất
#nếu value bằng nhau in key nhỏ hơn
max_val=0
result=None
for key,value in count.items():
    if value > max_val:
        max_val = value
        result = key
    elif value == max_val:
        if key < result:
            result = key
print(result)





```

[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
một dictionary mới bằng cách đảo ngược key và value của dictionary ban

```python
a=input().split()
dic={}
for i in range(0,len(a)-1,2):
    dic[int(a[i+1])]=a[i]

print(dic)
```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
mới

```python
a=input().split()
k=int(input())

dic={}
for i in range(0,len(a)-1,2):
    dic[a[i]]=a[i+1]
dic2={}
for key,value in dic.items():
    if int(value)>k:
        dic2[key]=int(value)

print(dic2)
        
```
