### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
def dien_tich(a,b,c):

    p = (a+b+c)/2
    s = (p*(p-a)*(p-b)*(p-c))**0.5
    return s
a,b,c = map(float, input().split())
print (dien_tich(a,b,c))
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def vi_tri(x, n):
    # kiểm tra nếu x là chữ cái hoa
    if 'A' <= x <= 'Z' :
        # vtri hiện tại- vtri A để đưa về dải 0-25
        # sau đó cộng thêm n rồi chia lấy phần dư cho 26 để vòng lại nếu quá Z
        # số bé hơn 26 khi chia lấy phần dư vẫn là chính nó
        return chr(((ord(x)-ord('A')) + n) % 26 + ord('A'))
    # kiểm tra nếu x là chữ cái thường 
    elif 'a' <= x <= 'z':
        return chr(((ord(x)-ord('a')) + n) % 26 + ord('a'))
    else:
        return x
print(vi_tri('a', 3))
# ord : chuyển chữ cái sang số thứ tự trong bảng mã ASCII 
# chr : chuyển ngược lại     
    


```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def chu_thuong(x):
    if len(x) != 1:
        raise ValueError
    if 'A' <= x <= 'Z':
        return chr(ord(x)+ 32)# cộng 32 vì bảng mã ASCII khi cộng thêm 32 sẽ từ hoa sang thường
    else:
        return x
print(chu_thuong('N'))
    
```
