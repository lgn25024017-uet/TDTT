Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 
kiểu số nguyên, đơn vị VND). 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): 
- Xe 4 chỗ: 11,000 VND 
- Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: 
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km 
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: 
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. 
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).



```python

def smart_taxi_fare(car_type,start_hour,km,wait_minutes):
    if car_type==4:
        mo_cua=11000
        gia_duoi_20=13000
        gia_tren_20=10500
    elif car_type==7:
        mo_cua=12000
        gia_duoi_20=14000
        gia_tren_20=12000
    else:
        return ('loại xe không phù hợp')
    
    if km<=0.8:
        tien_duong=mo_cua
    elif km<=20:
        tien_duong=mo_cua+(km-0.8)*gia_duoi_20
    else:
        tien_duong=mo_cua+19.2*gia_duoi_20+(km-20)*gia_tren_20
    
    if wait_minutes<=5:
        phi_cho=0
    else:
        minutes_charge=wait_minutes-5
        if minutes_charge/5==int(minutes_charge/5):
            phi_cho=(minutes_charge/5)*2000
        else:
            phi_cho=(int(minutes_charge/5)+1)*2000

    tam_tinh=tien_duong+phi_cho
    is_night= start_hour in [22,23,0,1,2,3,4,5]
    
    if is_night:
        total=tam_tinh*1.2
    else:
        total=tam_tinh

    if km>30 and not is_night:
        total*=0.9
    
    if total<50000 and km>5:
        total=50000
    
    return int(total)

print(f'Cước phí: {smart_taxi_fare(4,0,5000,5000):,} VND')


    


```

Yêu cầu: Cho năm số a, b, c, d, e, trả về True nếu tồn tại bất kỳ ba số nào trong đó bằng nhau; ngược lại trả về False. 
Không dùng tuple hay dictionary

```python
def check_three_equal(a, b, c, d, e):
    numbers = [a, b, c, d, e]
    
    for num in numbers:
        if numbers.count(num) >= 3:
            return True
            
    return False

print(check_three_equal(1, 2, 3, 1, 1))  
print(check_three_equal(1, 2, 3, 4, 5))  
```

Yêu cầu: Cho hai ngày sinh dưới dạng (day, month, year) của người 1 và người 2, trả về 1 nếu người 1 lớn tuổi hơn (ngày 
sinh sớm hơn), 2 nếu người 2 lớn tuổi hơn, và 0 nếu bằng nhau. 
Ví dụ: 27 2 2007 so với 28 2 2007 → trả về 1 vì ngày 27 sớm hơn nên người 1 lớn tuổi hơn.

```python
def so_sanh_tuoi(d1, m1, y1, d2, m2, y2):
    
    if y1 < y2: return 1
    if y1 > y2: return 2
    
    if m1 < m2: return 1
    if m1 > m2: return 2
    
    if d1 < d2: return 1
    if d1 > d2: return 2
    
    return 0
d1, m1, y1, d2, m2, y2=map(int,input().split())

print(so_sanh_tuoi(d1, m1, y1, d2, m2, y2))

```

Yêu cầu: Cho cân nặng (kg) và chiều cao (m) của hai người, hãy trả về người có BMI cao hơn. 
Trả về 1 nếu người 1 có BMI cao hơn, 2 nếu người 2 có BMI cao hơn, và 0 nếu bằng nhau. 
Nhắc lại: BMI = weight / (height height)

```python
def BMI(w1,h1,w2,h2):
    bmi1=w1/(h1**2)
    bmi2=w2/(h2**2)

   
    if bmi1 > bmi2:
        return 1
    elif bmi1< bmi2:
        return 2
    else:
        return 0
w1,h1=map(float,input().split())
w2,h2=map(float,input().split())
if BMI(w1,h1,w2,h2)==1:
    print('người 1 múp hơn 2')
elif BMI(w1,h1,w2,h2)==2:
    print('người 2 múp hơn 1')
else:
    print('hai người bmi bằng nhau')
```

```python
def tong_binh_phuong(n):
    tong = 0
    for i in range(1, n + 1):
        tong += i**2
    return tong

n = int(input())

print(f"Tổng bình phương các số từ 1 đến {n} là: {tong_binh_phuong(n)}")

```

Viết hàm trả về hệ số nhị thức C(n, k)

```python
def nhi_thuc(n,k):
    if k>n and k<0:
        return 0
    else:
        tich=1
        for i in range(1,n+1):
            tich*=i
        tich1=1
        for j in range(1,k+1):
            tich1*=j
        tich2=1
        for x in range(1,(n-k)+1):
            tich2*=x
    kqua=tich/(tich1*tich2)
    return int(kqua)

n,k=map(int,input().split())
print(nhi_thuc(n,k))
```

Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4).

```python
def dem_so_k(n):
    dem = 0
    moc_duoi = (n - 1) / 2
    for k in range(1, n):
        if k > moc_duoi:
            dem += 1
    return dem

n=int(input())
print(dem_so_k(n))
```
